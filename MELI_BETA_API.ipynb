```python
import json
import pandas as pd
import requests
```

```python
# Función de extracción de datos desde la API de Mercado Libre
def extract_data():
    url = 'https://api.mercadolibre.com/sites/MLA/search'
    params = {'q': 'Toyota Etios', 'category': 'MLA1743'}  # MLA1743 es el ID de la categoría de autos
    response = requests.get(url, params=params)
    data = response.json()
    return data
```

```python
def Transform():
    # Extracción de datos
    data = extract_data()
    site_id = data["site_id"]
    country_default_time_zone = data["country_default_time_zone"]
    query = data["query"]
    paging = data["paging"]
    results = data["results"]

    # Crear un dataframe vacío
    df = pd.DataFrame()

    # Acceder a los campos de cada resultado en la lista de resultados y almacenar los datos en el dataframe
    for aux in results:
        try:
            id_r = aux["id"]
            title = aux["title"]
            condition = aux["condition"]
            thumbnail_id = aux["thumbnail_id"]
            catalog_product_id = aux["catalog_product_id"]
            listing_type_id = aux["listing_type_id"]
            permalink = aux["permalink"]
            buying_mode = aux["buying_mode"]
            price = aux["price"]
            currency_id = aux["currency_id"]

            # Verificar si los campos están vacíos
            condicion_item = aux["attributes"][0]["value_name"] if aux["attributes"][0] else None
            unico_dueno = aux["attributes"][1]["value_name"] if aux["attributes"][1] else None
            control_traccion = aux["attributes"][2]["value_name"] if aux["attributes"][2] else None
            aire_acondicionado = aux["attributes"][3]["value_name"] if aux["attributes"][3] else None
            cilindros = aux["attributes"][4]["value_name"] if aux["attributes"][4] else None
            cilindrada = aux["attributes"][5]["value_name"] if aux["attributes"][5] else None
            capacidad_personas = aux["attributes"][6]["value_name"] if aux["attributes"][6] else None
            marca = aux["attributes"][7]["value_name"] if aux["attributes"][7] else None
            puertas = aux["attributes"][8]["value_name"] if aux["attributes"][8] else None
            motor = aux["attributes"][9]["value_name"] if aux["attributes"][9] else None
            tipo_combustible = aux["attributes"][10]["value_name"] if aux["attributes"][10] else None
            kilometros = aux["attributes"][11]["value_name"] if aux["attributes"][11] else None
            modelo = aux["attributes"][12]["value_name"] if aux["attributes"][12] else None
            transmision = aux["attributes"][13]["value_name"] if aux["attributes"][13] else None
            version = aux["attributes"][14]["value_name"] if aux["attributes"][14] else None
            #ano = aux["attributes"][15]["value_name"] if aux["attributes"][15] else None

            ciudad = aux["seller_address"]["city"]["name"]
            estado = aux["seller_address"]["state"]["name"]
            pais = aux["seller_address"]["country"]["name"]

            data_dict = {
                "id": id_r,
                "title": title,
                "condition": condition,
                "thumbnail_id": thumbnail_id,
                "catalog_product_id": catalog_product_id,
                "listing_type_id": listing_type_id,
                "permalink": permalink,
                "buying_mode": buying_mode,
                "price": price,
                "currency_id": currency_id,
                "condicion_item": condicion_item,
                "unico_dueno": unico_dueno,
                "control_traccion": control_traccion,
                "aire_acondicionado": aire_acondicionado,
                "cilindros": cilindros,
                "cilindrada": cilindrada,
                "capacidad_personas": capacidad_personas,
                "marca": marca,
                "puertas": puertas,
                "motor": motor,
                "tipo_combustible": tipo_combustible,
                "kilometros": kilometros,
                "modelo": modelo,
                "transmision": transmision,
                "version": version,
                #"ano": ano,
                "ciudad": ciudad,
                "estado": estado,
                "pais": pais
            }

            df = df.append(data_dict, ignore_index=True)

        except Exception as e:
            print(f"Error en el procesamiento de un elemento: {e}")
            continue

    return df
```

```python
def load(df):
    file_path = 'output.csv'
    df.to_csv(file_path, index=False)
```

```python
df = Transform()
load(df)
```
